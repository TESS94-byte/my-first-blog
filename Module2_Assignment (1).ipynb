```python
#4.1
secret = 3
guess = 10
if guess < secret:
    print('too low')
elif guess > secret:
    print('too high')
else:
    print('just right')
```

```python
#4.2
small = True
green = True
if not small and not green :
    print('pumpkin')
elif small and not green :
    print('cherry')
elif not small and green :
    print('watermelon')
else:
    print('pea')
```

```python
#6.2
guess_me = 7
number = 1
while True:
    if number < guess_me :
        print('too low')
    elif number == guess_me :
        print('found it !')
        break
    else:
        print('oops')
        break
    number += 1
```
