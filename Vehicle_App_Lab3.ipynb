```python
#Name: Tessnime zaoui ben salem
# File: Vehicle_App_Lab3.ipynb
#Description: This program creates a Vehicle superclass and an Automobile subclass.
#It asks the user to enter information about a car(year, make, model, number of doors,
#and type of roof), then displays the information in a clean format.
#Variables:
#vehicle_type(str): Type of vehicle
#year(str): Year of the vehicle entered by the user
#make (str): Manufacturer of the vehicle 
#model(str): Model name of the vehicle
#doors(str): Number of doors
#roof (str): Type of roof 


#superclass:
class Vehicle:
    def __init__(self, vehicle_type):
        self.vehicle_type = vehicle_type
#subclass:
class Automobile(Vehicle):
    def __init__(self, vehicle_type, year, make, model, doors, roof):
        super().__init__(vehicle_type)
        self.year = year
        self.make = make
        self.model = model
        self.doors = doors
        self.roof = roof
#main Program
vehicle_type = "car"
year = input("Enter vehicle year")
make = input("Enter vehicle make")
model = input("Enter vehicle model")
doors = input("Enter numbers of doors 2 or 4 ")
roof = input("Enter type of the roof solid or sun roof")
car = Automobile(vehicle_type, year, make, model, doors, roof)
print("\n ***** Vehicle Information ***** ")
print('Vehicle type: ', car.vehicle_type)
print('Year: ', car.year)
print('Make: ', car.make)
print("Model: ", car.model)
print("Number of doors: ", car.doors)
print("Type of roof: ", car.roof)
```
